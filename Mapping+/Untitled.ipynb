```python
import folium
import pandas
map = folium.Map(location = [38.58, -99.09], zoom_start = 6, tiles = "Stamen Terrain")

vol_cd = pandas.read_csv("Volcanoes.txt")
lat = list(vol_cd["LAT"])
lon = vol_cd["LON"]
loc = [lat,lon]
vol_cd
```

```python
fg = folium.FeatureGroup(name = "My Map")
for coordinates in loc:
  fg.add_child(folium.Marker(location = loc, popup = "Hi I am a Marker", icon = folium.Icon(color = 'green')))
map.add_child(fg)

map.save("Map1.html")
```
